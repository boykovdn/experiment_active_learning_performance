```python
import graphviz
```

```python
g = graphviz.Digraph('al_experiment', format='svg')

g.node('dataset')
g.node('classifier')
g.node('**kwargs', label='T=10\nbatch_size=20')
g.node('binary_classification_scores', shape='box')
g.node('gp_al_q', label='gp_al; random', shape='box')
g.node('gp_al_r', label='gp_al; active', shape='box')
g.node('score_active; mean, error')
g.node('score_random; mean, error')

g.edge('dataset', 'binary_classification_scores')
g.edge('classifier', 'binary_classification_scores')
g.edge('**kwargs', 'binary_classification_scores')
g.edge('binary_classification_scores', 'gp_al_q', label='1...N')
g.edge('binary_classification_scores', 'gp_al_r', label='1...N')
g.edge('gp_al_q', 'score_active; mean, error')
g.edge('gp_al_r', 'score_random; mean, error')

g.render(directory='/notebooks')
g
```
